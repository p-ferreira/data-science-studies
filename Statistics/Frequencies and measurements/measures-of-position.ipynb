## <font color = green> Quartiles, deciles and percentiles </font>
***

There are a number of position measures similar in their design to the median, although they are not measures of central tendency. As is known, the median divides the distribution into two equal parts in terms of the number of elements in each part. Quartiles, on the other hand, allow the distribution to be divided into four equal parts in terms of the number of elements in each; the deciles in ten parts and the centiles in one hundred equal parts.

```python
import pandas as pd
import seaborn as sns

data = pd.read_csv('data/data.csv', sep = ',')


data['Income'].quantile([0.25, 0.5, 0.75])
```

```python
decilis = [i/10 for i in range(1, 10)]
decilis
```

```python
data['Income'].quantile(decilis)
```

```python
percentiles = [i/100 for i in range(1, 100)]
data['Income'].quantile(percentiles)
```

```python
ax = sns.distplot(data['Age'],
                  hist_kws= {'cumulative': True},
                  kde_kws={'cumulative': True},
                  bins = 10)
ax.figure.set_size_inches(14,6)
ax.set_title('Accumulated frequency distribution', fontsize = 18)
ax.set_ylabel('Accumulated', fontsize = 14)
ax.set_xlabel('Years', fontsize = 14)
ax
```

```python
data['Age'].quantile([i / 10 for i in range(1,10)])
```

## <font color = green> Box-plot </font>
***

The box plot gives an idea of the position, dispersion, asymmetry, tails and outliers. The central position is given by the median and the dispersion by $ IIQ $. The relative positions of $ Q1 $, $ Median $ and $ Q3 $ give a sense of the symmetry of the distribution. Tail lengths are given by the lines that go from the rectangle to the remote values and by the outliers.

```python
ax = sns.boxplot(x = 'Height', data = data, orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Height', fontsize = 18)
ax.set_xlabel('Metters', fontsize = 14)
ax
```

```python
ax = sns.boxplot(x = 'Height', y='Sex', data = data, orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Height', fontsize = 18)
ax.set_xlabel('Metters', fontsize = 14)
ax
```

```python
ax = sns.boxplot(x = 'Income', data = data.query('Income < 10000'), orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Income', fontsize = 18)
ax.set_xlabel('R$', fontsize = 14)
ax
```

```python
formatted_sexSeries = data['Sex'].replace({0: 'Man', 1: 'Woman'})

formatted_data = data
formatted_data['Sex'] = formatted_sexSeries

ax = sns.boxplot(x = 'Income', y = 'Sex', data = formatted_data.query('Income < 10000'), orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Income', fontsize = 18)
ax.set_xlabel('R$', fontsize = 14)
ax
```

```python
ax = sns.boxplot(x = 'Years of study', 
                 data = data,
                 orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Years of study', fontsize = 18)
ax.set_xlabel('Years', fontsize = 14)

ax
```

```python
ax = sns.boxplot(x = 'Years of study', y = 'Sex', data = formatted_data, orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Years of study', fontsize = 18)
ax.set_xlabel('Years', fontsize = 14)
ax
```

```python
colorLabels = { 0: 'Indigenous',
          2: 'White',
          4: 'Black',
          6: 'Yellow',
          8: 'Brown',
          9: 'No declaration'}

formatted_color_series = data['Color'].replace(colorLabels)

formatted_data_byColor = data
formatted_data_byColor['Color'] = formatted_color_series

formatted_data_byColor
ax = sns.boxplot(x = 'Income', y = 'Color', data = formatted_data_byColor.query('Income < 5000'), orient = 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Income', fontsize = 18)
ax.set_xlabel('R$', fontsize = 14)
ax
```
