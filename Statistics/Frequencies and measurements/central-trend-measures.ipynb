***

It is represented by $\mu $ when referring to the population and by $\bar {X} $ when referring to the sample

$ n $ = number of observations (records)

$ X_i $ = value of the i-th observation (record)

## <font color = green>Median </font>
***

To obtain the median of a data set, we must proceed as follows:
1. Sort the data set;
2. Identify the number of observations (records) in the data set ($ n $);
3. Identify the middle element:

> When $ n $ is odd, the position of the median element will be obtained as follows:


# $$ Element_ {Md}=\frac {n + 1} 2 $$

> When $ n $ is even, the position of the median element will be obtained as follows:


# $$ Element_ {Md}=\frac {n} 2 $$

4. Obtain the median:

> When $ n $ is odd:


# $$ Md = X_{Element_{Md}} $$

> When $ n $ is even:


# $$ Md = \frac{X_ {Element_{Md}} + X_{Element_{Md} +1}} 2 $$
***

### Evaluating the variable INCOME

```python
import seaborn as sns

ax = sns.distplot(data.query('Income < 20000')['Income'])
ax.figure.set_size_inches(12, 6)
ax
```

```python
mode = data['Income'].mode()[0]
mode
```

```python
median = data['Income'].median()
median
```

```python
mean = data['Income'].mean()
mean
```

```python
mode < median < mean
```

### Evaluating the variable HEIGHT

```python
ax = sns.distplot(data['Height'])
ax.figure.set_size_inches(12, 6)
ax
```

```python
heightSeries = data['Height']
print(heightSeries.mean())
print(heightSeries.median())
print(heightSeries.mode())
```

### Evaluating the variable YEARS OF STUDY

```python
yearsOfStudySeries = data['Years of study']
ax = sns.distplot(yearsOfStudySeries, bins = 17)
ax.figure.set_size_inches(12, 6)
ax
```

```python
print(yearsOfStudySeries.mode())
print(yearsOfStudySeries.median())
print(yearsOfStudySeries.mean())
```
