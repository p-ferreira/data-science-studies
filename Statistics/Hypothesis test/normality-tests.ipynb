# <font color=green>Hypothesis tests</font>
***

Statistical tests are decision rules that allow to evaluate the reasonableness of the hypotheses made on population parameters and accept them or reject them as probably true or false based on a sample.

## <font color=green>Normality Tests</font>
***

### Importing libraries

```python
from scipy.stats import normaltest
```

The * normaltest * function tests the null hypothesis $H_0$ that the sample comes from a normal distribution.

### Defining the test significance ($\alpha$)

```python
significance = 0.05
```

### Testing the Income variable

```python
data['Income'].hist(bins = 50)
```

### <font color = 'red'> $ p $ value criterion </font>

> ### Reject $H_0$ if $p\leq 0.05$

```python
stat_test, p_value = normaltest(data['Income'])
print(stat_test, p_value)
```

```python
p_value <= significance
```

### Testing the Height variable

```python
data['Height'].hist(bins = 50)
```

### <font color ='red'>$p$ value criterion </font>

> ### Reject $H_0$ if $p\leq 0.05$

```python
stat_test, p_value = normaltest(data['Height'])
print(stat_test, p_value)
```

```python
p_value <= significance
```

## <font color=green>Basic steps of a test</font>
***

### ** Step 1 ** - formulation of the hypotheses $H_0$ and $H_1$;

> ### <font color='red'> Important points </font>
> - In general, the target of the study should be formulated as the alternative hypothesis$H_1$.
> - The null hypothesis always affirms an equality or population property, and$H_1$the inequality that negates$H_0$.
> - In the case of the null hypothesis$H_0$the equality can be represented by a simple equality "$=$" or by "$\geq$" and "$\leq$". Always complementary to that established by the alternative hypothesis.
> - The alternative hypothesis$H_1$must define an inequality that can be a simple difference "$\neq$" or of the types "$>$" and "$<$".


### ** Step 2 ** - choose the appropriate sample distribution;

> ### <font color = 'red'> Important points </font>
> - When the sample size has 30 elements or more, the normal distribution should be used, as established by the ** central limit theorem **.
> - For a sample size less than 30 elements, and if we can say that the population is distributed approximately as a normal and the population standard deviation is known, the normal distribution should be used.
> - For a sample size less than 30 elements, and if we can say that the population is distributed approximately as a normal and the population standard deviation is unknown, Student's t-distribution should be used.

### ** Step 3 ** - fixing the significance of the test ($\ alpha$), which defines the regions of acceptance and rejection of the hypotheses (the most frequent values ​​are 10%, 5% and 1%);

> ### <font color = 'red'> Important points </font>
> - The ** confidence level ** ($1 - \ alpha$) represents the probability of correctness of the estimate. In a complementary way, the ** significance level ** ($\ alpha$) expresses the probability of error of the estimate.
>
> ![Confidence level and significance](https://caelum-online-public.s3.amazonaws.com/1229-estatistica-parte3/01/img001.png)
>
> - The ** confidence level ** represents the degree of reliability of the estimate result being within a certain range. When we set a ** 95% confidence level ** in a survey, for example, we are assuming that there is a 95% probability that the survey results represent reality well, that is, they are correct.
>
>![Acceptance and Rejection Areas](https://caelum-online-public.s3.amazonaws.com/1229-estatistica-parte3/01/img002.png)

### ** Step 4 ** - calculation of the test statistic and verification of this value with the areas of acceptance and rejection of the test;

> ### <font color = 'red'> Important points </font>
> - In parametric tests, the relative distance between the sample statistic and the alleged value as probable.
> - In this step, the sample statistics necessary to perform the test are obtained (mean, standard deviation, degrees of freedom, etc.)


### ** Step 5 ** - Acceptance or rejection of the null hypothesis.

> ### <font color = 'red'> Important points </font>
> - If the acceptance range contains the test statistic,$H_0$is accepted as statistically valid and$H_1$is rejected as such.
> - If the acceptance range does not contain the test statistic,$H_0$is rejected and$H_1$is accepted as probably true.
> - Acceptance is also verified with the tail probability (p-value): if greater than$\ alpha$,$H_0$is accepted.
