```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))

data = pd.read_csv('data/aluguel.csv', sep = ';')
data.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
g1.scatter(data['Valor'], data['Area'])
g1.set_title('Value X Area')

g2.hist(data['Valor'])
g2.set_title('Histogram')

data_g3 = data['Valor'].sample(100)
data_g3.index = range(len(data_g3))
g3.plot(data_g3)
g3.set_title('Sample (Value)')

group = data.groupby('Tipo')['Valor']
label = group.mean().index
values = group.mean().values
g4.bar(label, values)
g4.set_title('Mean Value by type')

```
