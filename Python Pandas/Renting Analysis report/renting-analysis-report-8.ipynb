## Identifying and removing outliers

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

plt.rc('figure', figsize = (20,10))

data = pd.read_csv('data/residential_rent.csv', sep = ';')
```

```python
data.boxplot(['Valor'])
```

```python
data[data['Valor'] >= 500000]
```

```python
value = data['Valor']

q1 = value.quantile(.25)
q3 = value.quantile(.75)

IIQ = q3 - q1

inferior_limit = q1 - 1.5 * IIQ
superior_limit = q3 + 1.5 * IIQ

```

```python
selection = (value >= inferior_limit) & (value <= superior_limit)
new_data = data[selection]

new_data.boxplot(['Valor'])
```

```python
data.hist(['Valor'])
new_data.hist(['Valor'])
```

```python
data.boxplot(['Valor'], by = [
    'Tipo'
])
```

```python
group_type = data.groupby('Tipo')['Valor']
type(group_type)
```

```python
group_type.groups
```

```python
q1 = group_type.quantile(.25)
q3 = group_type.quantile(.75)

IIQ = q3 - q1

inferior_limit = q1 - 1.5 * IIQ
superior_limit = q3 + 1.5 * IIQ
```

```python
new_data = pd.DataFrame()
for _type in group_type.groups.keys():
    is_type = data['Tipo'] == _type
    is_in_limit = (data['Valor'] >= inferior_limit[_type]) & (data['Valor'] <= superior_limit[_type])
    selection = is_type & is_in_limit    
    selection_data = data[selection]
    new_data = pd.concat([new_data, selection_data])

new_data.boxplot(['Valor'], by = ['Tipo'])
```

```python
new_data.to_csv('data/residential_rent_without_outliers.csv', sep = ';', index = False)
```
