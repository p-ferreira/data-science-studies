```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
import warnings

data = pd.read_csv('data/beer_consumption.csv', sep = ';')
# warnings.filterwarnings('ignore') #ignores all warnings
warnings.filterwarnings(action='once')
```

# <font color = 'red' style = 'font-size: 30px;'> Dependent Variable X Explanatory Variables (pairplot) </font>
<hr style = 'border: 2px solid red;'>

## Scatter plots between variables in the dataset

## seaborn.pairplot

https://seaborn.pydata.org/generated/seaborn.pairplot.html?highlight=pairplot#seaborn.pairplot

<p style = 'font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; '> Plots the relationship between pairs of variables in a dataset. </p>

```python
ax = sns.pairplot(data)
```

## Plotting the pairplot fixing only one variable on the y axis

```python
ax = sns.pairplot(data, y_vars='consumption', x_vars=['min_temp', 'mean_temp', 'max_temp', 'rain', 'weekend'], diag_kind=None)
ax.fig.suptitle('Dispersion between variables', fontsize = 20, y= 1.15)
```

```python
ax = sns.pairplot(data, y_vars='consumption', x_vars=['min_temp', 'mean_temp', 'max_temp', 'rain', 'weekend'], diag_kind=None, kind='reg')
ax.fig.suptitle('Dispersion between variables', fontsize = 20, y= 1.15)
ax
```

# <font color = 'red' style = 'font-size: 30px;'> Dependent Variable X Explanatory Variables (jointplot) </font>
<hr style = 'border: 2px solid red;'>

## seaborn.jointplot

https://seaborn.pydata.org/generated/seaborn.jointplot.html?highlight=jointplot#seaborn.jointplot

<p style = 'font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; '> Plots the relationship between two variables and their respective frequency distributions. </p>

```python
ax = sns.jointplot(x='max_temp', y='consumption', data=data)
ax.fig.suptitle('Dispersion - Consumption X Temperatue', fontsize = 18, y=1.05)
ax.set_axis_labels('Max temperature', 'Beer consumption', fontsize = 14)
ax
```

## Plotting a jointplot with the estimated regression line

```python
ax = sns.jointplot(x='max_temp', y='consumption', data=data, kind='reg')
ax.fig.suptitle('Dispersion - Consumption X Temperatue', fontsize = 18 , y=1.05)
ax.set_axis_labels('Max temperature', 'Beer consumption', fontsize = 14)
ax
```

# <font color = 'red' style = 'font-size: 30px;'> Dependent Variable X Explanatory Variables (lmplot) </font>
<hr style = 'border: 2px solid red;'>

### seaborn.lmplot

https://seaborn.pydata.org/generated/seaborn.lmplot.html?highlight=lmplot#seaborn.lmplot

<p style = 'font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; '> Plots the regression line between two variables together with the dispersion between them. </p>

```python
ax = sns.lmplot(x='max_temp', y='consumption', data=data)
ax.fig.suptitle('Regression line - Consumption X Temperature', fontsize=16, y=1.02)
ax.set_xlabels('Max temperature', fontsize=14)
ax.set_ylabels('Beer consumption (liters)', fontsize=14)
ax
```

```python
ax = sns.lmplot(x='max_temp', y='consumption', data=data, hue='weekend', markers=['o', '*'], legend=False)
ax.fig.suptitle('Regression line - Consumption X Weekend', fontsize=16, y=1.02)
ax.set_xlabels('Max temperature', fontsize=14)
ax.set_ylabels('Beer consumption (liters)', fontsize=14)
ax.add_legend(title = 'Is weekend?')
ax
```

```python
ax = sns.lmplot(x='max_temp', y='consumption', data=data, col='weekend')
ax.fig.suptitle('Regression line - Consumption X Weekend', fontsize=16, y=1.02)
ax.set_xlabels('Max temperature', fontsize=14)
ax.set_ylabels('Beer consumption (liters)', fontsize=14)
ax
```
