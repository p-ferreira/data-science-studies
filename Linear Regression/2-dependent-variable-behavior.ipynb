```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np

data = pd.read_csv('data/beer_consumption.csv', sep = ';')
```

# <font color = 'red' style = 'font-size: 30px;'> Dependent Variable Behavior (Y) </font>
<hr style = 'border: 2px solid red;'>

# Graphical analysis

## Plotting the * dependent * (y) variable
https://pandas.pydata.org/pandas-docs/stable/visualization.html

```python
fig, ax = plt.subplots(figsize=(20, 6))

ax.set_title('Beer consumption', fontsize=20)
ax.set_ylabel('Liters', fontsize=16)
ax.set_xlabel('Days', fontsize=16)
ax = data['consumption'].plot(fontsize = 14)
```

## Importing seaborn library
https://seaborn.pydata.org/
<p style = 'font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify; '> Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing statistical graphs. </p>

```python
import seaborn as sns
```

## Box plot of the * dependent * variable (y)

https://seaborn.pydata.org/generated/seaborn.boxplot.html?highlight=boxplot#seaborn.boxplot

```python
ax = sns.boxplot(data=data['consumption'], orient='v', width=0.2)
ax.figure.set_size_inches(12, 6)
ax.set_title('Beer consumption', fontsize = 20)
ax.set_ylabel('Liters', fontsize = 16)
ax
```

# <font color = 'red' style = 'font-size: 30px;'> Box Plot with Two Variables </font>
<hr style = 'border: 2px solid red;'>

## Investigating the * dependent * variable (y) according to a given characteristic

```python
ax = sns.boxplot(y='consumption', x='weekend', data=data, orient='v', width=0.2)
ax.figure.set_size_inches(12, 6)
ax.set_title('Beer consumption', fontsize = 20)
ax.set_ylabel('Liters', fontsize = 16)
ax.set_xlabel('Weekend', fontsize = 16)
ax
```

## * Seaborn * library style and color settings

### Style control

> #### API
> https://seaborn.pydata.org/api.html#style-api

> #### Tutorial
> https://seaborn.pydata.org/tutorial/aesthetics.html#aesthetics-tutorial

### Color palette

> #### API
> https://seaborn.pydata.org/api.html#palette-api

> #### Tutorial
> https://seaborn.pydata.org/tutorial/color_palettes.html#palette-tutorial

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')

ax = sns.boxplot(y='consumption', x='weekend', data=data, orient='v', width=0.2)
ax.figure.set_size_inches(12, 6)
ax.set_title('Beer consumption', fontsize = 20)
ax.set_ylabel('Liters', fontsize = 16)
ax.set_xlabel('Weekend', fontsize = 16)
ax
```

# <font color = 'red' style = 'font-size: 30px;'> Frequency Distribution </font>
<hr style = 'border: 2px solid red;'>

## Frequency distribution of the * dependent * (y) variable

https://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=distplot#seaborn.distplot

```python
ax = sns.distplot(data['consumption'])
ax.figure.set_size_inches(12, 6)
ax.set_title('Frequency distribution', fontsize = 20)
ax.set_ylabel('Beer consumption (liters)', fontsize = 16)
ax
```
