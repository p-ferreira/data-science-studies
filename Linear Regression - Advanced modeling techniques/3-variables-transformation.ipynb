## Importing numpy library

```python
import numpy as np
```

## Applying the logarithmic transformation to * dataset * data

https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.log.html

```python
np.log(0)
```

```python
data['log_value'] = np.log(data['value'])
data['log_area'] = np.log(data['area'])
data['log_dist_beach'] = np.log(data['dist_beach'] + 1) #adds one due to log(0) = -inf
data['log_dist_pharmacy'] = np.log(data['dist_pharmacy'] + 1) #adds one due to log(0) = -inf
```

```python
data.head()
```

## Frequency distribution of the * transformed dependent * variable (y)

```python
ax = sns.distplot(data['log_value'])
ax.figure.set_size_inches(15, 6)
ax.set_title('Frequency distribution', fontsize=20)
ax.set_xlabel('log of real state price', fontsize=16)
ax
```

## Scatter plots between the transformed variables of the dataset

```python
ax = sns.pairplot(data, y_vars = 'log_value', x_vars=['log_area', 'log_dist_beach', 'log_dist_pharmacy'], diag_kind=None, height=5)
ax.fig.suptitle('dispersion between transformed variables', fontsize=20, y=1.05)
ax
```
